```python
# Carga del dataset
import pandas as pd
df = pd.read_csv("meteo.csv")
print(len(df))
df
```

```python
## Vemos cuales variables presentan una falta de datos
df.isnull().sum()
```

```python
## Preparacion del dataset

# Para este modelo, no se utilizarÃ¡n las variables cuyo % de ausencia sea mayor a 50 ya que al no haber sufientes daots, puede que
# afecten la predicciÃ²n del modelo de manera inesperada. (cual serÃ¡ el valor ideal?)
# Se elimina Location para tratar de generalizarlo a otros lugares

df = df.drop(columns = ['Location', 'Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm', 'RISK_MM'])
df.head()
```

```python
# Eliminar las filas cuyos valores categoricos no se encuentren presentes
df = df.dropna(subset=["WindGustDir"])
df = df.dropna(subset=["WindDir9am"])
df = df.dropna(subset=["WindDir3pm"])
df = df.dropna(subset=["RainToday"])

# Se llenan los valores numericos faltantes con el promedio de las columnas
df["MinTemp"].fillna(df["MinTemp"].mean(), inplace=True)
df["MaxTemp"].fillna(df["MaxTemp"].mean(), inplace=True)
df["Humidity9am"].fillna(df["Humidity9am"].mean(), inplace=True)
df["Humidity3pm"].fillna(df["Humidity3pm"].mean(), inplace=True)
df["Pressure9am"].fillna(df["Pressure9am"].mean(), inplace=True)
df["Pressure3pm"].fillna(df["Pressure3pm"].mean(), inplace=True)
df["Temp9am"].fillna(df["Temp9am"].mean(), inplace=True)
df["Temp3pm"].fillna(df["Temp3pm"].mean(), inplace=True)


df.isnull().sum()
```

```python
# Transformacion de la fecha en dia del aÃ±o. (por las estaciones)
a = pd.to_datetime(df["Date"], format='%Y-%m-%d').dt.dayofyear

# Tambien se transforman las variables categoricas a numericas
b = pd.get_dummies(df['WindGustDir'], prefix = "WGD")
c = pd.get_dummies(df['WindDir9am'], prefix = "WD9")
d = pd.get_dummies(df['WindDir3pm'], prefix = "WD3")
e = pd.get_dummies(df['RainToday'], prefix = "RT")

# Ejemplo de como queda la fecha
a[:5]
```

```python
# Concatenacion de los datos
frames = [a, b, c, d, e, df]
df = pd.concat(frames, axis = 1)
df = df.drop(columns = ['Date', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday'])
df.head()
```

```python
X = df.iloc[0:len(df),:-1]
y = df.iloc[0:len(df),len(df.columns)-1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=25)

from sklearn import tree

#Entrenamiento
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
#Predicciones
y_pred = clf.predict(X_test)

#EvaluaciÃ³n del rendimiento del clasificador

from sklearn.metrics import confusion_matrix
confusion_matrix_tree = confusion_matrix(y_test, y_pred)

print()
print("Matriz de confusion")
print(confusion_matrix_tree)

#Print de la matriz de confusiÃ³n

from sklearn.metrics import classification_report
print()
print("Recall y precision")
print(classification_report(y_test, y_pred))
```

```python
## Modelo utilizando regresion logistica
from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression()
LogReg.fit(X_train, y_train)

print("coeficientes del modelo: "+str(LogReg.coef_))
print("intercept: "+str(LogReg.intercept_))

#Predicciones
y_pred = LogReg.predict(X_test)

#EvaluaciÃ³n del rendimiento del clasificador
from sklearn.metrics import confusion_matrix
confusion_matrix_log = confusion_matrix(y_test, y_pred)

print()
print("Matriz de confusion")
print(confusion_matrix_log)

print()
print("Recall y precision")
print(classification_report(y_test, y_pred))
```

```python
# Finalmente, podemos ver que el modelo que utiliza la regresion logistica como algoritmo funciona mejor que el desicion tree 
# para este problema

accu_tree = (confusion_matrix_tree[0][0]+confusion_matrix_tree[1][1])/(confusion_matrix_tree[0][0] + confusion_matrix_tree[1][1] + confusion_matrix_tree[1][0]+confusion_matrix_tree[0][1])

accu_log = (confusion_matrix_log[0][0]+confusion_matrix_log[1][1])/(confusion_matrix_log[0][0] + confusion_matrix_log[1][1] + confusion_matrix_log[1][0]+confusion_matrix_log[0][1])

print("Accuracy Tree: " + str(accu_tree))
print("Accuracy Log: " + str(accu_log))
```
